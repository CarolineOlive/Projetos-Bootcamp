### Regi√£o do Pa√≠s com maior n√∫mero de casos

Uma forma ter uma vis√£o mais clara dos dados apontados pelo gr√°fico 'N√∫mero total de interna√ß√£o por c√¢ncer de mama e do colo do √∫tero por Unidade Federativa - 2008 a 2020' √© agrupar os estados por regi√£o.

```
#Transformando a coluna 'Unidade da Federa√ß√£o' em indice para us√°-lo na classifica√ß√£o por regi√£o.

dados_usaveis = dados_sort.set_index("Unidade da Federa√ß√£o")
dados_usaveis.head()
```

```
# Criando uma nova coluna para associar uma regi√£o a cada estado.
# A partir do indice, foi selecionado apenas o primeiro n√∫mero para usar como refer√™ncia para as regi√µes
dados_usaveis["Regi√£o"] = dados_usaveis.index.str[0]

# criando um dicion√°rio que associa a chave 'index'(estados) √†s correspondentes regi√µes. 
regiao_estado = {1:"Norte", 2:"Nordeste", 3:"Sudeste", 4:"Sul", 5:"Centro-Oeste"}

# criando uma fun√ß√£o que associa o estado (key) √† regi√£o (value) e cria a nova coluna.
dados_usaveis["Regi√£o"] = dados_usaveis["Regi√£o"].map(lambda x: regiao_estado[int(x)])

dados_usaveis["Regi√£o"]
```

```
dados_usaveis.head()
```

```
# Somando os numeros de interna√ß√µes dos estados por regi√£o
norte = dados_usaveis.query("Regi√£o == 'Norte'")["Total"].sum()
sul = dados_usaveis.query("Regi√£o == 'Sul'")["Total"].sum()
sudeste = dados_usaveis.query("Regi√£o == 'Sudeste'")["Total"].sum()
centro_oeste = dados_usaveis.query("Regi√£o == 'Centro-Oeste'")["Total"].sum()
nordeste = dados_usaveis.query("Regi√£o == 'Nordeste'")["Total"].sum()
```

```
# criando um dicion√°rio para associar regi√£o e n¬∫ de interna√ß√£o
dados_regiao = {'Regi√£o': ["Norte", "Sul", "Sudeste", "Centro-Oeste", "Nordeste"], 'Total de Interna√ß√µes': [norte, sul, sudeste, centro_oeste, nordeste]}
```

```
# Criando um novo DataFrame com a regi√£o e o total de interna√ß√µes
dados_por_regiao = pd.DataFrame(data=dados_regiao)
dados_por_regiao
```

#### Valores absolutos

```
regioes = dados_por_regiao.sort_values("Total de Interna√ß√µes", ascending=False)
regioes.plot(x="Regi√£o", y="Total de Interna√ß√µes", kind='bar', figsize = (9, 7), color = 'mediumvioletred')
plt.title("N√∫mero total de interna√ß√£o por c√¢ncer de mama e do colo do √∫tero por Regi√£o do Brasil - 2008 a 2020")
plt.legend("")
plt.show()
```

Como indicado no gr√°fico anterior, olhando para os dados por regi√£o, vemos que a regi√£o Norte √© a que apresenta o menor n√∫mero de interna√ß√µes, sendo quase 13 vezes menor que o da regia√ß√£o sudeste.

Quando observamos apenas o n√∫mero absoluto de interna√ß√µes, sem considerar fatores como o n√∫mero de habitantes por regi√£o/unidade federativa, corremos o risco de interpretar erroneamente os dados. Um estado com um menor n√∫mero de interna√ß√µes e, tamb√©m, de habitantes pode ter, proporcionalmente, o mesmo n√∫mero de interna√ß√µes que um estados com maior n√∫mero de habitantes e interna√ß√µes.

Para verificar esse fator, vamos avaliar o n√∫mero de habitantes por regi√£o do pa√≠s.

```
regioes
```

```
# criando uma nova coluna com o n√∫mero de habitantes de cada regi√£o
regioes['Habitantes'] = [89012240, 57374243, 30192315, 18672591, 16504303]
regioes
```

#### Propor√ß√£o por milh√£o de habitantes

```
# para mostrar a propor√ß√£o de interna√ß√µes por habitantes de uma forma mais clara, optei por transformar os valores em n¬∫ de interna√ß√µes por milh√£o de habitantes.
internacoes_por_milhao = (regioes["Total de Interna√ß√µes"] * 1000000) / regioes["Habitantes"]
```

```
# usei a fun√ß√£o round para arredondar os valores obtidos
internacoes_por_milhao = round(internacoes_por_milhao, 0) 
```

```
# criando uma nova coluna com o n√∫mero de habitantes de cada regi√£o
regioes["Interna√ß√µes por Milh√£o de habitantes"] = internacoes_por_milhao
regioes
```

```
regioes["Percentual"] = regioes["Total de Interna√ß√µes"] / regioes['Habitantes']
regioes
```

```
regioes = regioes.sort_values("Interna√ß√µes por Milh√£o de habitantes", ascending=False)
regioes.plot(x="Regi√£o", y="Interna√ß√µes por Milh√£o de habitantes", kind='bar', figsize = (9, 7), color = 'deeppink')
plt.title("N√∫mero de interna√ß√µes por milh√£o de habitantes")
plt.legend("")
plt.show()
```

Em n√∫meros totais, a regi√£o sul estava em 3¬∞ lugar em n¬∫ de interna√ß√µes. No entanto, proporcionalmente, √© a regi√£o com maior n√∫mero de interna√ß√µes. J√° a regi√£o Norte, continua sendo a regi√£o com menor n√∫mero.
